```python
import os
import re
import numpy as np

Path = "E:\\GPImage\\HandleImage\\TrainingSet\\"
EnhancePath = 'E:\\GPImage\\HandleImage\\Train_Enhance_Image\\'
```

```python

grayList = os.listdir(Path)
enhanceList = os.listdir(EnhancePath)
grayList.sort()
enhanceList.sort()

gray_totalNum = len(grayList)
enhance_totalNum = len(enhanceList)
```

```python
gray_name = []
for i in range(gray_totalNum):
    name = str(grayList[i])
    num_name = re.sub("\D","", name)
    gray_name.append(num_name)
# gray_name.sort()
print(gray_name)
```

```python
enhance_name = []
for i in range(enhance_totalNum):
    name = str(enhanceList[i])
    num_name = re.sub("\D","", name)
    enhance_name.append(num_name)
# enhance_name.sort()
print(enhance_name)
```

```python
gray = np.argsort(gray_name)
enhance = np.argsort(enhance_name)

print(gray)
print("-----------------")
print(enhance)
```

```python

for i in range(gray_totalNum):
#     获取enhance的文件名
    name = EnhancePath + str(enhance_name[i]) + '.jpg'
#     获取gray对应的文件名
    re_name = EnhancePath + str(grayList[gray[i]])
    print(name)
    print(re_name)
    print("-------------------")
#     name_path = Path + "\\\\" + str(i) + ".png"
#     re_name_path = Path + "\\\\" + str(i) + "_left.png"
    os.renames(name, re_name)
print("文件重命名完成")
```
