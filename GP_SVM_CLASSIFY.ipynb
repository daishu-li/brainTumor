```python
# 导入所需的包
import tensorflow as tf
import os
import numpy as np
from io import BytesIO
import pickle
import cv2 as cv
import h5py
import matplotlib.pyplot as plt
import skimage
from skimage.transform import rotate
from skimage.feature import local_binary_pattern
from skimage import data, io,data_dir,filters, feature
from sklearn.svm import LinearSVC
from sklearn.externals import joblib
from skimage.color import label2rgb
```

```python
hog_train_Path = 'E:\\GPImage\\HandleImage\\Train_Enhance_Image\\'
classify_Image_Path = 'E:\\GPImage\\HandleImage\\EnhanceImage\\'
```

```python
# 提取HOG特征
def get_hog_feature(hog_train_Path, hog_list, hog_num):
    features = []
    labels = []
    for i in range(hog_num):
        # print("第" + str(i) + "次输出")
        i_hog_filePath = hog_train_Path + hog_list[i]
        
        image = cv.imread(i_hog_filePath, cv.COLOR_BGR2GRAY)
        # print(str(image.shape))
        if(image.shape == (256,256)):
            
            #  添加 文件标签
            # 获取文件名
            y = hog_list[i]
            #     print(y)

            # 去除文件后缀
            y = os.path.splitext(y)
            #     print(y)
            y_ = y[0]
            #     print(y_)

            # 去除文件中的数字，获取文件标签
            y_ = "".join(filter(str.isalpha, y_))
            #     print(y_)
            # 添加标签  
            labels.append(y_)
            
            
            # 添加图像hog的特征数据
            a = np.array(65536)
            a.flags.writeable = True
            a = np.copy(image)
            a.shape = 256, 256
            a = a.T
            i_feature = feature.hog(a, 
                                  orientations=9,
                                  pixels_per_cell=(32, 32),
                                  cells_per_block=(4, 4),
                                  block_norm = 'L1',
                                  transform_sqrt=True,
                                  feature_vector=True,
                                  visualise=False)
             #print(str(i_feature.shape))
             # plt.plot(feature)
            # plt.show()
            features.append(i_feature)
            
    return features,labels
```

```python
# 获取HOG特征
hog_list = os.listdir(hog_train_Path)
hog_num = len(hog_list)
hog_features, hog_labels = get_hog_feature(hog_train_Path, hog_list, hog_num)

print(hog_features)
print(hog_labels)

clf = LinearSVC()
clf.fit(hog_features, hog_labels)
joblib.dump(clf, "E:\\GPImage\\HandleImage\\digits_cls.pkl", compress=3)
print("运算完成！！！！！！！！！")
```
