```python
import numpy as np
import os
import cv2 as cv
import matplotlib.pyplot as plt
import matplotlib.image as mp

from math import sqrt
from collections import Counter

from sklearn import neighbors
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report
```

```python
trainPath = 'E:\\GPImage\\HandleImage\\TrainingSet\\'
testPath = 'E:\\GPImage\\HandleImage\\GrayImage\\'
savePath = 'E:\\GPImage\\HandleImage\\HandleImage\\'
```

```python
X = []
Y = []

fileList = os.listdir(trainPath)
totalNum = len(fileList)
# print(fileList[i])

# 遍历图片
for i in range(0, totalNum):
    iPath = trainPath + str(fileList[i])
#     print(iPath)
#  添加 文件标签
    # 获取文件名
    y = fileList[i]
#     print(y)
    
    # 去除文件后缀
    y = os.path.splitext(y)
#     print(y)
    y_ = y[0]
#     print(y_)
    
    # 去除文件中的数字，获取文件标签
    y_ = "".join(filter(str.isalpha, y_))
#     print(y_)
    
    # 添加标签  
    Y.append(y_)
    
#   添加文件图片数据 
    src = cv.imread(iPath)
    image = cv.cvtColor(src, cv.COLOR_BGR2GRAY)

#     显示图片与标签验证
#     plt.imshow(image, cmap='gray')
#     plt.axis('off')
#     plt.show()
#     print(image)

    X.append(image)
    
X_train = np.array(X)
Y_train = np.array(Y)
```

```python

distances = []
testList = os.listdir(testPath)
testNum = len(testList)
# print(testList)

for i in range(1,testNum):
    img_path = testPath + str(testList[i])
    print(img_path)
    
    
    src = cv.imread(img_path)
    image = cv.cvtColor(src, cv.COLOR_BGR2GRAY)
    
    classCount = {}
    
    xPath = testPath + str(testList[i])
    xSRC = cv.imread(xPath)
    x = cv.cvtColor(xSRC, cv.COLOR_BGR2GRAY)
    
    distances = [sqrt(np.sum(x_train - x)**2) for x_train in X_train]
    
#     for x_train in X_train:
#         d = sqrt(np.sum(x_train - x)**2)
#         distances.append(d)
    
    sortedDistance = np.argsort(distances)
    
#     print(sortedDistance)
#     sortedDistance[::-1]
#     print('---------------------\n')
#     print(sortedDistance)
    
    labels = []
    for k in range(28):
        index = sortedDistance[k]
#         print(index)
        label = Y_train[index]
        labels.append(label)

    # 获取出现次数最多的邻居的标签
    data = Counter(labels)
    key = data.most_common(1)[0][0]
    
    # 给测试集的图片进行分类
    if key == 'up':
        path = savePath + 'UP\\'
        name = path + str(i) + '_up.png'
        cv.imwrite(name, image)            
    elif key == 'eye':
        path = savePath + 'EYE\\'
        name = path + str(i) + '_eye.png'
        cv.imwrite(name, image)        
    elif key == 'back':
        path = savePath + 'BACK\\'
        name = path + str(i) + '_back.png'
        cv.imwrite(name, image)       
    elif key == 'left':
        path = savePath + 'LEFT\\'
        name = path + str(i) + '_left.png'
        cv.imwrite(name, image) 
    else:
        path = savePath + 'FACE\\'
        name = path + str(i) + '_face.png'
        cv.imwrite(name, image)
        
    print('--------------------------------------------------------------------------------------------------')
   
```
